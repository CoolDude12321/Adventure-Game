```python
# This code represents a text based simple adventure game where the player searches for hidden treasures.
# The code uses Github Copilot & VSCode as the IDE.

# Defining the main function to start the game
def start_game():
    print("Welcome to the Treasure Hunt Adventure!")
    print("Your aim in this game is to find hidden treasures.")
    player_name = input("Before we strt the game, please enter your name: ")
    print(f"Welcome, {player_name}! Let's begin this adventure. Hold tight!")

    while True:
        print("\nYou are at a crossroads. Where would you like to go?")
        print("1. Would you like to explore the dark magical forest?")
        print("2. Would you like to enter the ancient abandoned castle?")
        choice = input("Enter 1 or 2: ")

        if choice == '1':
            return forest_path()
        elif choice == '2':
            return cave_path()
        else:
            print("Invalid choice. Please choose 1 or 2.")

# Defining the forest path function
def forest_path():
    print("\n You step into the dense, dark forest. The air is damp, and eerie sounds echo around you.")
    print("You come across two possible actions:")
    print("1. Follow the sound of a flowing river")
    print("2. Climb a tall tree to get a better view")

    choice = input("Enter 1 or 2: ")

    if choice == '1':
        print("\n You follow the river and discover a sparkling chest half-buried in the sand!")
        print(" CONGRATULATIONS! You found the legendary treasure and WON the adventure!")
    elif choice == "2":
        print("\n You climb the tree, but the branch snaps and you fall to the ground.")
        print("You are injured and unable to continue your journey.")
        print(" You lost the adventure. Better luck next time.")
    else:
        print("Invalid choice. A wild animal attacks unexpectedly!")
        print(" You lost the adventure.")


# Defining the castle path function
def cave_path():
    print("\n You enter a dark and echoing cave. It's cold and silent.")
    print("Two options appear:")
    print("1. Light a torch")
    print("2. Proceed in total darkness")

    choice = input("What will you do? (1/2): ")

    if choice == "1":
        print("\n You light a torch and see ancient markings on the walls…")
        print("They guide you to a hidden chamber containing a glittering treasure chest!")
        print(" You found the treasure and WON the game! ")
    elif choice == "2":
        print("\n You stumble in the darkness and fall into a deep pit.")
        print(" You lost the adventure.")
    else:
        print("Invalid choice. Rocks fall from the ceiling and block the exit.")
        print(" Game Over.")


def play_again(): 
    choice = input("\nWould you like to play again? (yes/no): ").lower()
    return choice == "yes"


# Run the game loop
while True:
    start_game()
    if not play_again():
        print("\nThanks for playing Adventure Quest! Goodbye ")
        break  
```
